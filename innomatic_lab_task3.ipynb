Problem Statement 1:Employee Performance Bonus Eligibility

```python
# Employee performance dictionary
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
```

```python

#Find the highest performance score
highest_score = max(employees.values())

```

```python
#Find all employees with the highest score
top_performers = []

for name in employees:
    if employees[name] == highest_score:
        top_performers.append(name)
```

```python
# Step 3: Display result
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {highest_score})")
```

Problem statement 2 : Search Query Keyword Analysis


```python
# Input sentence
query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

# Remove punctuation 
import string
for p in string.punctuation:
    query = query.replace(p, "")

#Split sentence into words
words = query.split()

# Count frequency of each keyword
frequency = {}

for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

# Display words appearing more than once
result = {}

for word in frequency:
    if frequency[word] > 1:
        result[word] = frequency[word]

print(result)
```

Problem statement:3:Sensor Data Validation

```python
# Sensor readings list
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# List to store valid readings
valid_readings = []

# Loop through list with index
for hour, value in enumerate(sensor_readings):
    if value % 2 == 0:   # check if even
        valid_readings.append((hour, value))

# Display result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

Problem statement :4:Email Domain Usage Analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Step 1: Count domains
domain_count = {}

for email in emails:
    domain = email.split("@")[1]   # get domain part
    
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

# Step 2: Calculate percentage
total_users = len(emails)

for domain in domain_count:
    percentage = (domain_count[domain] / total_users) * 100
    print(f"{domain}: {int(percentage)}%")
```

Problem Statement 5:Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: Calculate average sales
average_sales = sum(sales) / len(sales)

# Step 2: Find spike days (30% above average)
threshold = average_sales * 1.30

for i, value in enumerate(sales):
    if value > threshold:
        print(f"Day {i+1}: {value}")
```

Problem Statement 6: Duplicate User ID Detection

```python
# List of user IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Step 1: Count occurrences of each user ID
count_ids = {}

for uid in user_ids:
    if uid in count_ids:
        count_ids[uid] += 1
    else:
        count_ids[uid] = 1

# Step 2: Display duplicates only
for uid in count_ids:
    if count_ids[uid] > 1:
        print(f"{uid} → {count_ids[uid]} times")
```
