Problem 1: Mobile Recharge Validation System


```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]
    
    while True:
        amount = int(input("Enter recharge amount: ₹"))
         #Recharge amount must be at least 50
        if amount < 50:
            print("Recharge amount must be at least ₹50. Please try again.\n")
            continue
        # Check if amount is a valid plan
        if amount in valid_plans:
            print("Recharge successful! Plan activated.Thank you!")
            break
        else:
            print("Invalid plan selected. Available plans:", valid_plans)
            print("Please try again.\n")

validate_recharge()     # Call the function
```

Problem 2: Inventory Reorder Alert System

```python
def check_inventory(products):
    
    for product, quantity in products.items():
        
        if quantity < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")
# Example dictionary
inventory = {
    "Mobile": 100,
    "Laptop": 12,
    "Headphones": 15,
    "Charger": 102
}

check_inventory(inventory) # Call function
```

 Problem 3: Student Result Processing System


```python
def process_result(marks):
    #Calculate total using loop
    total = 0
    for mark in marks:
        total += mark
    #Calculate average
    average = total / len(marks)
    #Decide pass or fail
    if average >= 50:
        result = "Pass"
    else:
        result = "Fail"
    # Print results
    print("Average Marks:", average)
    print("Result:", result)
    
    return result
    
student_marks = [96, 85, 80, 100]  #input

process_result(student_marks)  # Call function
```

Problem 4: Cab Fare Estimator with Retry Option


```python
def calculate_fare(distance, is_peak):
    base_fare = 50
    per_km_rate = 15
    fare = base_fare + (distance * per_km_rate)
 # Adding 20% extra during peak hours
    if is_peak:
        fare += fare * 0.20
    
    return fare

# Retry system using while loop
while True:
    distance = float(input("Enter distance in km: "))
    peak_input = input("Is it peak hour? (yes/no): ").lower()
    is_peak = True if peak_input == "yes" else False
    total_fare = calculate_fare(distance, is_peak)
    print("Total Fare: ₹", total_fare)
    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        print("Thank you for using Cab Fare Estimator!")
        break
```

Problem 5: Employee Attendance Eligibility Checker


```python
def check_attendance(attendance):
    # Count present days using loop
    present_count = 0
    for day in attendance:
        if day == "P":
            present_count += 1
    #Calculate percentage
    total_days = len(attendance)
    percentage = (present_count / total_days) * 100
    #Check eligibility
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

attendance_list = ["P", "P", "A", "P", "P", "A", "P", "P"]  #example input

result = check_attendance(attendance_list)     # Call function

print("Attendance Percentage:", (attendance_list.count("P") / len(attendance_list)) * 100)
print("Status:", result)
```

Problem 6: Password Strength Checker


```python
def check_password(password):
# Minimum length
    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"
    
    has_digit = False
    has_special = False
    #Check digit and special character
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True
    
    if has_digit and has_special:
        return "Strong Password "
    else:
        return "Weak Password (Must contain at least one digit and one special character)"
    
user_password = input("Enter password: ")
result = check_password(user_password)
print(result)
```
