Problem 1: Smart Light Controller


```python
class SmartLight:
    
    def __init__(self, name):
        self.name = name
        self.status = "OFF"
    
    def control(self, action):
        self.status = action
        print(self.name, "is", self.status)


# Main program
name = input("Enter Light Name: ")
action = input("Enter Action (ON/OFF): ")

light = SmartLight(name)
light.control(action)
```

Problem 2: Employee ID Card System


```python
class Employee:
    
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department
    
    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


#main program 
name = input("Employee Name: ")
emp_id = input("Employee ID: ")
department = input("Department: ")

emp = Employee(name, emp_id, department)
emp.display_id_card()
```

Problem 3: Mobile Contact Record


```python
class Contact:
    
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    
    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


#main program
name = input("Contact Name: ")
phone = input("Phone Number: ")

person = Contact(name, phone)
person.display_contact()
```

Problem 4: Product Price Tag Generator


```python
class Product:
    
    def __init__(self, name, price):
        self.name = name
        self.price = price
    
    def display_price_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))


#main program 
product_name = input("Product Name: ")
product_price = input("Price: ")

item = Product(product_name, product_price)
item.display_price_tag()
```

Problem 5: Movie Rating Display System


```python
class Movie:
    
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating
    
    def display_details(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")

#main program
movie_name = input("Movie Name: ")
movie_rating = input("Rating: ")

film = Movie(movie_name, movie_rating)
film.display_details()
```

Problem 6: Delivery Address Manager


```python
class Delivery:
    
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address
    
    def display_delivery_details(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)

#main program
name = input("Customer Name: ")
address = input("Address: ")

order = Delivery(name, address)
order.display_delivery_details()
```
