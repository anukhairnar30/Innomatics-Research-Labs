```python
cuisine_stats = (
    df.groupby("cuisine")
      .agg(
          restaurant_count=("restaurant_id", "nunique"),
          total_revenue=("total_amount", "sum")
      )
      .sort_values(["restaurant_count", "total_revenue"], ascending=[True, False])
)

cuisine_stats
```

```python
df.columns
```

```python
gold_percentage = round(
    (df[df["membership"] == "Gold"].shape[0] / df.shape[0]) * 100
)

gold_percentage
```

```python
restaurant_stats = (
    df.groupby("restaurant_id")
      .agg(
          order_count=("total_amount", "count"),
          avg_order_value=("total_amount", "mean")
      )
)
```

```python
filtered_restaurants = restaurant_stats[
    restaurant_stats["order_count"] < 20
]
```

```python
top_restaurant_id = filtered_restaurants["avg_order_value"].idxmax()
top_restaurant_id
```

```python
restaurants.loc[
    restaurants["restaurant_id"] == top_restaurant_id,
    "restaurant_name"
]
```
